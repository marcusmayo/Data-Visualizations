```
#This is a bar graph showing the number of huricanes occurring in the United States along the Atlantic Coast.
#The year is plotted on the x-axis and the number of hurricanes occuring as the y-axis.
import pandas as pd
import plotly.graph_objects as go

#I create the data frame from the github csv file.
df=pd.read_csv("https://raw.githubusercontent.com/marcusmayo/Data_Visualizations/master/Hurricanes.csv")

#Here I identify the x and y axis; and the legend of the bar graph.
fig=go.Figure(go.Bar(x=df['Year'], y=df["Hurricanes"], name='US Hurricanes along the Atlantic coast'))

#I create the parameters for the title of the bar graph before creating the bar graph.
fig.update_layout(title='US Hurricanes along the Atlantic coast', plot_bgcolor='rgb(230,230,230)', showlegend=True)

#Show the created chart
fig.show()
```

```
#This program is a histogram plot of the temperatures over 2014 and 2015 for San Francisco and Moscow.
import matplotlib.pyplot as plt
import pandas as pd

#Create the data frame using the pd library
df=pd.read_csv('https://raw.githubusercontent.com/marcusmayo/Data_Visualizations/master/CityTemps.csv')

#I create the parameters of the chart utilizing different colors for the different cities.
plt.hist(df['Moscow'], color='blue', edgecolor='black', bins=int(180/5))
plt.hist(df['San Francisco'], color='green', edgecolor='black', bins=int(180/5))

#Here I create the labels for the x and y axis before creating the histogram.
plt.title('City Temps')
plt.xlabel('Temps')
plt.ylabel('No. of Days')
plt.show()
```

```
#This program is a pie chart showing the percentage of car models released by every manufacturer.
#I import the necessary libraries needed
import pandas as pd
import plotly.graph_objects as go #Plotly is an open source graphing library

#Create the data frame with the github data csv provided
df=pd.read_csv('https://raw.githubusercontent.com/marcusmayo/Data_Visualizations/master/Cars2015.csv')

#Create a defined function to create a pie chart with the data frame
def pie_count(data, field='Make'):
  print("Creating Pie Chart by ", field, " count..........................")
  data[field]=data[field].replace('', 'NA') #Replace blanks with NA(Not Available)
  data=data[field].value_counts().to_frame() #Create a new data frame with count of make
  labels=data.index.tolist() #List of make names
  datavals=data[field].tolist() #List of model count
  title='No. of Models by Manufacturer'
  trace=go.Pie(labels=labels, values=datavals)
  layout=go.Layout(
      title=title,
      height=600,
      margin=go.layout.Margin(l=0, r=200, b=100, t=100, pad=4)) #Margins - left, right, top, bottom, padding
  fig=go.Figure(data=[trace], layout=layout)
  fig.show()

#Plot the pie chart using the defined function against the data frame
pie_count(df)
```

```
#This program is to create a simple plot with labels
#Import matplotlab library
import matplotlib.pyplot as plt

#Define plot size i n inches (width, height) $ resolution(DPI)
fig=plt.figure(figsize=(4, 5), dpi=100)

#Define some data
x=[1,2,3,4]
y=[20,21,20.5,20.8]

#Define error data
y_error=[0.12,0.13,0.2,0.1]

#Plot data
plt.plot(x, y, linestyle='dashed', marker='o', color='green')

#Plot only error bars
plt.errorbar(x, y, yerr=y_error, linestyle='None', marker='None', color='red')

#Configure x-axes
plt.xlim(0.5, 4.5)
plt.xticks([1,2,3,4])

#Configure y-axes
plt.ylim(19.8, 21.2)
plt.yticks([20,21,20.5,20.8])

#Labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

#Title
plt.title('Simple Plot')

#Adjust plot
plt.subplots_adjust(left=0.18)

#Define font size
plt.rc('font', size=14)

#Show plot
plt.show()
```

```
#This program is to create a scatter graph of 50 random values on the x and y axis
#Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

#Fixing random state for reproducibility
np.random.seed(19680801)
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
area=(30*np.random.rand(N))**2 #0 to 15 point radii

#Print scatter plot
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

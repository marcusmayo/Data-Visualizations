```
#This program is to plot the total sales per month for the year 2011
#Importing pandas
import pandas as pd

#Create the data frame
df=pd.read_csv('https://raw.githubusercontent.com/marcusmayo/Data-Sources/master/BigMartSalesData.csv', index_col='Month', low_memory=False)

#Selecting data
sales1=df[['Year', 'Amount', 'Quantity']]

#Filter the data
sales2=sales1[sales1['Year']==2011]

#Aggregate the data
groupby_sum1=sales2.groupby(['Month', 'Year']).sum()

#Print the data frame with the Quantity column in ascending order
print(groupby_sum1.sort_values('Quantity', ascending=True))
```

```
#This program is to use the data frame above and plot sales per month as a Bar Chart
import matplotlib.pyplot as plt

#The data frame was grouped by month and plotted against the amount as bar chart
groupby_sum1.groupby('Month').sum().plot(y='Amount', kind='bar')

#Print the bar graph
plt.show()
```

```
#This program is to create a pie chart by country using the data frame above
import matplotlib.pyplot as plt

#The original data fram was grouped by country and pltted against the amount as a pie chart
df.groupby('Country').sum().plot(y='Amount', kind='pie')

#I moved the pie chart legend so it would not overlap the pie chart
plt.legend(loc=(1.3,0))

#Print the pie chart
plt.show()
```

```
#I did not like the pie chart visulaization so I changed it to a horizontal bar chart
import matplotlib.pyplot as plt

#The original data frame was grouped by country against the amount as a horizontal bar chart
df.groupby('Country').sum().plot(y='Amount', kind='barh')

#I increased the bar chart size for better visability
fig_size=plt.rcParams['figure.figsize']
fig_size[0]=20
fig_size[1]=10
plt.rcParams['figure.figsize']=fig_size

#Show the bar graph
plt.show()
```

```
#
import numpy as np #Numpy provides fast mathematical function processing
import pandas as pd #Pandas is used for easy indexing and data collection tools
import matplotlib as mpl #Machine learning libraries
import matplotlib.pyplot as plt #Library for data visualizations
import seaborn as sns #import library to create a heatmap
%matplotlib inline
```

```
#This is how yopu name the dataset
#I named it NYCdata
#pd.read_csv is how you open a csv file
NYCdata=pd.read_csv("https://raw.githubusercontent.com/TERRENCE2019/Thinkful/master/nycrollingsales.csv")
```

```
NYCdata.isnull().values.any()
#This code checks to tell if any data in dataframe has NULL/NaN values
#True means yes there are Null or NAN values
#Fals means No there ar not any Null or NAN values
```

```
NYCdata2=NYCdata.dropna()
#Code to remove all (NAN)not a number / Null values
```

```
NYCdata2.isnull().values.any()
#False means No there are not any Null or NAN values
```

```
NYCdata2
```

```
#The code doesn't pick up on the NAN value so we have to manually fix it (ETL)
NYCdata2['SALE PRICE']=NYCdata2['SALE PRICE'].str.replace(r'-','')
```

```
NYCdata2['BUILDING CLASS CATEGORY']=NYCdata2['BUILDING CLASS CATEGORY'].str.replace(r'-','')
```

```
#Lets remove entire column
del NYCdata2['SALE DATE']
```

```
NYCdata2
```

```
#This is the code for sizing the heatmap
f, ax = plt.subplots(figsize=(15,10))
plt.title('US Data Correlations')

#Draw the heatmap using seaborn
sns.heatmap(NYCdata2.corr(),linewidths=0.15,vmax=1.0, square=True, cmap='YlGnBu', linecolor='black', annot=True)
b, t = plt.ylim() #discover the values for bottom and top
b += 0.75 #Add 0.5 to the bottom
t -= 0.75 #Subtract 0.5 from the top
plt.ylim(b, t) # Update the ylim(bottom,top) values
plt.show() #ta da!
```

```
#This is the code to remove a column
#We rename the cell so the new cell data can be within the new dataframe
NYCdata3=NYCdata2.drop(['Unnamed: 0'], axis=1)
```

```
#This is the code for sizing the heatmap
f, ax = plt.subplots(figsize=(15,10))
plt.title('US Data Correlations')

#Draw the heatmap using seaborn
sns.heatmap(NYCdata3.corr(),linewidths=0.15,vmax=1.0, square=True, cmap='YlGnBu', linecolor='black', annot=True)
b, t = plt.ylim() #discover the values for bottom and top
b += 0.75 #Add 0.5 to the bottom
t -= 0.75 #Subtract 0.5 from the top
plt.ylim(b, t) # Update the ylim(bottom,top) values
plt.show() #ta da!
```
